# A Decade of Best Sellers

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
df = pd.read_csv('books.csv')
```

## Jeff Kinney, author of 'Diary of a Wimpy Kid' has the most best sellers

```python
authors = pd.DataFrame(df.Author.value_counts().reset_index().values, columns=["Author", "Count"])[:10]
sns.barplot('Count', 'Author', data=authors)
print()
```

```python
sns.barplot('Year', 'Reviews', data=df, ci=None)
print()
```
