# Twitter: Trumps Platform for Bullying

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
sns.set()
df = pd.read_csv('trump.csv')
df['date'] = pd.to_datetime(df['date'])
```

## Trump loves to target the media

```python
targets = pd.DataFrame(df.target.value_counts().reset_index().values, columns=["Target", "Count"])
sns.barplot('Count', 'Target', data=targets[:20])
print()
```

Democrats, Hillary Clinton, and Joe Biden are notable top targets

## So, what does Trump think of the media?

```python
media_insults = df[df["target"] == 'the-media']
media_insults = " ".join(media_insults['insult'].values)
wc = WordCloud().generate(media_insults)
plt.imshow(wc)
plt.axis("off")
plt.show()
```

This clearly demonstraits that Trumps loves to call the media "fake" and point out "fake news"

## And what does Trump think of his political opponents?

```python
joe_hilary = df[(df["target"] == 'joe-biden') | (df["target"] == 'hillary-clinton')][['date', 'target', 'tweet']]
jh_counts = joe_hilary.groupby(['target', 'date']).agg({'tweet': "count"})
sns.lineplot(x='date', y='tweet', hue='target', data=jh_counts)
print()
```

Trump tweets against political opponents dramatically increase around election season. Interestingly, Trump attacked Hillary Clinton far more in 2016 than Joe Biden in 2020. 

## Summary

Trump loves to use Twitter to target opponents and critics. He has a lot of time and energy insulting the media for being 'fake'. When he's not tweeting about the media, he's targetting his political opponents-- spreading harsh insults around election season. Trumps time in office has shown us how Twitter can be used to bully others.
